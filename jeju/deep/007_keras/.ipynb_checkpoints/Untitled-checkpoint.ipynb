```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('iris.csv', header=None)
dataset = df.values
X = dataset[1:,0:4]
Y = dataset[1:,4]
X,Y
```

```python
# 이번엔 회귀를 해보자
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from keras.models import Sequential
from keras.layers import Dense, Dropout
import pandas
import numpy
```

```python
def baseline_model():
    model = Sequential()
    # 506x13, 13x13 => 506x13
    model.add(Dense(6, input_dim=4, kernel_initializer='normal', activation='relu' ))
    # 506x13, 13x1 =. 506x1
    model.add(Dense(3,kernel_initializer='normal'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model 
```

```python
seed = 7
numpy.random.seed(seed)
estimator = KerasClassifier(build_fn = baseline_model, nb_epoch=100, batch_size=5, verbose=0)
```

```python
kfold = KFold(n_splits=3, random_state=seed)
results = cross_val_score(estimator, X, Y, cv=kfold)
print(results)
```

```python
numpy.random.seed(seed)
estimators= []
estimators.append(('standardize', StandardScaler()))
estimators.append(('mlp', KerasClassifier(build_fn = baseline_model, epochs=50, batch_size=5, verbose=0)))
# Pipeline 매개변수가 리스트를 요구
pipeline = Pipeline(estimators)
kfold =KFold(n_splits=3, random_state=seed)
results = cross_val_score(pipeline, X, Y, cv=kfold)
print(results)
```

```python
pipeline.fit(X,Y)
```

```python
res = pipeline.predict(X)
res
```
