```python
import numpy as np
def make_data(N, err=1.0,rseed=1):
    rng=np.random.RandomState(rseed)
    X=rng.rand(N,1)**2
    y = 10-1/(X.ravel() + 0.1)
    if err > 0:
        y += err * rng.randn(N)
    return X,y
X,y = make_data(40)
print(type(X))
```

```python
from sklearn.model_selection import GridSearchCV # parameter tuning
from sklearn.preprocessing import PolynomialFeatures # 다차원특성 축소
from sklearn.linear_model import LinearRegression
import numpy as np

# 파이프라인 리텀
def PolynomialRegression(degree=2, **kwargs): # dict 변동매개변수
    return make_pipeline(PolynomialFeatures(degree), LinearRegression(**kwargs))

# 0~20
param_grid={'polynomialfeatures__degree':np.arange(21),
           'linearregression__fit_intercept':[True, False],
           'linearregression__normalize':[True, False]}
```

```python
# cross validation
grid = GridSearchCV(PolynomialRegression(), param_grid, cv=10)
grid
```

```python
grid.fit(X,y)
```

```python
grid.best_params_ # 비선형회귀 4차일때 좋은 결과
```

```python
grid.best_estimator_ # 가장 좋은 파라미터로 만든 모델
```

```python
grid.best_score_ # 가장 좋은 점수
```

```python
import matplotlib.pyplot as plt
model =grid.best_estimator_

X_test=np.linspace(-0.1,1.1,500)[:,None] # 한 열로 만들어주기 위해서

# 차수 일치
plt.scatter(X.ravel(), y) # 평평하게 1차원
lim = plt.axis()

# 예측
y_test=model.fit(X,y).predict(X_test)
plt.plot(X_test.ravel(), y_test)
plt.axis(lim)
plt.show()
```

```python
import numpy as np
import pandas as pd

dataset= [10,12,12,13,12,11,14,13,15,10,10,10,100,12,14,13, 12,10, 10,11,12,15,12,13,12,11,14,13,15,10,15,12,10,14,13,15,10] 
outliers=[]

def detect_outliter(data_1):
    threshold=3
    mean_1=np.mean(data_1) # 평균
    std_1 = np.std(data_1) # 표준편차
    for y in data_1:
        z_score = (y - mean_1) / std_1 # z-점수
        if np.abs(z_score) > threshold: # 문지방 : 경계값
            outliers.append(y)
    return outliers

outlier_datapoints = detect_outliter(dataset)
print(outlier_datapoints)
```

```python
import pandas as pd

boston = load_boston() # data(독립변수), target(종속변수), feature_names(변수?>)

data = pd.DataFrame(boston.data)
data.columns = boston.feature_names


data['PRICE'] = boston.target # 하나의 데이터로 묶음
print(data.info())
print(data.describe())

import xgboost as xgb
from sklearn.metrics import mean_squared_error # 평가
```

```python
# 데이터프레임은 ndarray + dict(순서를 보장, 중복을 허용)
X,y = data.iloc[:,:-1], data.iloc[:,-1] # 추가된 데이터
data_dmatrix = xgb.DMatrix(data=X, label=y) # 전용행렬 자기들만의 전용 포맷?

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=123)

# reg:linear 선형 회귀
# learning_rate 경사하강법에 등장
# optimization을 찾기위해
# max_depath 과적합 방지
# n_estimators=10
xg_reg=xgb.XGBRegressor(objective='reg:linear', colsample_bytree=0.3, learning_rate=0.1,max_depath=5,alpha=10,n_estimators=10)

xg_reg.fit(X_train, y_train)
preds=xg_reg.predict(X_test) #  y_bar 예측치
rmse=np.sqrt(mean_squared_error(y_test, preds))
print('RMSE : %f'%(rmse))

xgb.plot_tree(xg_reg, num_trees=0)
plt.rcParams['figure.figsize'] = [500,200]
plt.show()
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score
```

```python
dataset = loadtxt('pima.data', delimiter=',')
X = dataset[:,0:8]
Y = dataset[:,8]
seed=7

test_size=0.33
X_train, X_test, y_train, y_test=train_test_split(X,Y, test_size=test_size, random_state=seed)
model=XGBClassifier()
model.fit(X_train, y_train)
print(model)
```

```python
from xgboost import plot_importance
from matplotlib import pyplot
plot_importance(model)
pyplot.show()
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print('정확도 : %.2f%%'%(accuracy *100))
```

```python
import numpy as np
from sklearn.feature_selection import SelectFromModel
thresholds = np.sort(model.feature_importances_)
```

```python
for thresh in thresholds :
    selection= SelectFromModel(model, threshold=thresh, prefit=True)
    
    select_X_train=selection.transform(X_train)
    # 결계선이하의 중요성을 가진 변수 제거
    selection_model = XGBClassifier()
    selection_model.fit(select_X_train, y_train)
    select_X_test = selection.transform(X_test) # 테스트용데이터 변수
    y_pred = selection_model.predict(select_X_test)
    
    predictions = [round(value) for value in y_pred]
    accuracy= accuracy_score(y_test, predictions)
    print('Thresh=%.3f, n%d, Accuracy : %.2f%%'%(thresh, select_X_train.shape[1], accuracy*100))
    
```

```python
import pickle # 메모리의 내용을 그대로 저장
pickle.dump(model, open('pima.pickle.dat', 'wb'))
loaded_model = pickle.load(open('pima.pickle.dat', 'rb')) # binary

y_pred = loaded_model.predict(X_test)
predictions=[round(value) for value in y_pred]
accuracy = accuracy_score(y_test, predictions)
print('Accuracy : %.2f%%'%(accuracy*100))
```

```python
# 오늘 배운 것

# 1. pipeline  :   chaining 즉 앞에함수가 처리한 내용을 다음함수로 전달해주는것
#                 =앞에 전처리하고나서 그 다음 모델로 넘기는 것이라고 생각하자
#                   전처리하기 전에 train 와 test 모델을 분리해야하는데 명령어 2번쓰면 힘드니까 1번을 쓰기 위해 파이프로 묶는다
#               파이프에 들어가는 함수들에 이름을 달아줬었다 (언더바 2개 이용) 
# 2. GridSearchCV :  노가다 방지 /  parmater 튜닝을 하는 것 / parameter 조합을 만들어서 test 를 해보는 것이다!!
# 3. model 
#     3.1 decision tree : regressor, classfier 가 있다
#     3.2  RandomForest                                                                    
#     3.3 XGboost   ( 이걸이용해서 처리하려면 Dmatrix 를 만들었다, 
#                 --> 전용매트릭스를 만들어서 하는 방법이있고, 그거 없이 모델을 직접호출해서 하는법 2가지가 있다 ) 
#
```
