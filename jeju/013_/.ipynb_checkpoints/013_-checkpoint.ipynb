```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np

# 이미지를 파일로 출력하고 로딩한다음 글씨만 추출
def make_hello(N=1000, rseed=42):
    fig, ax = plt.subplots(figsize=(4, 1))
    fig.subplots_adjust(left=0, right=1, bottom=0, top=1)
    ax.axis('off')
    ax.text(0.5, 0.4, 'HELLO', va='center', ha='center', weight='bold', size=85)
    fig.savefig('hello.png')
    plt.close(fig)   

    from matplotlib.image import imread
    data = imread('hello.png')[::-1, :, 0].T   
    print("이미지차원", data.shape)
   
    print(data)
    rng = np.random.RandomState(rseed)
    X = rng.rand(4 * N, 2)
    print("만든 갯수",X.shape)
    print((X * data.shape).shape)
    i, j = (X * data.shape).astype(int).T
   
    mask = (data[i, j] < 1)
    X = X[mask]
    print("새로운X갯수", X.shape)
    print("원래이미지의 차수 ", data.shape)
    X[:, 0] *= (data.shape[0] / data.shape[1])
    X = X[:N]

    return X[np.argsort(X[:, 0])]
```

```python
X=make_hello(1000)
# rainbow에서 5개 선택
colorize = dict(c=X[:,0], cmap=plt.cm.get_cmap('rainbow', 5))

plt.scatter(X[:,0], X[:,1], **colorize)
plt.axis('equal')
```

```python
print(X.shape)
```

```python
def rotate(X,angle):
    theta=np.deg2rad(angle) # 라디안 - 호의 길이
    R=[[np.cos(theta), np.sin(theta)], [-np.sin(theta), np.cos(theta)]] # 2차원 행렬
    
    print(type(R))
    
    return np.dot(X,R)

X2 = rotate(X,20)+5 # 1000 X2 2X2 = >1000X2
plt.scatter(X2[:,0], X2[:,1], **colorize)
plt.axis('equal')
```

```python
# 각 점들간의 상호거리(euclidan distance)
from sklearn.metrics import pairwise_distances # 디폴트 = 유클리디안 거리

D = pairwise_distances(X) # 거리행렬
print(D.shape)
D[:5,:5]

plt.imshow(D, zorder=2, cmap='Blues', interpolation='nearest')
plt.colorbar()
```

```python
D2 = pairwise_distances(X2)
np.allclose(D,D2) # 원형을 유지하고있음
```

```python
# 다형체 manifold
from sklearn.manifold import MDS

# 미리 계산되어진 거리행렬을 이요해 차원 축소함
model=MDS(n_components=2, dissimilarity='precomputed', random_state=1)
out=model.fit_transform(D)
plt.scatter(out[:,0],out[:,1], **colorize)
plt.axis('equal')
print(out)
```

```python
# 고유값 분해
import numpy as np

rng = np.random.RandomState(10) # seed 값
C= rng.randn(3,3) # normal
print(np.dot(C,C.T)) # 전치 행렬, 행렬의 거듭제곱(정방행렬이면서 대칭행렬이 나온다.)

# 고유값 분해
e,V=np.linalg.eigh(np.dot(C,C.T))

print('eigenvector',V) # 고유벡터
print('eigenvalue',e) # 고유값

np.dot(V[1],V[2]) # 두벡터의 내적 => 직교 딱나오진않는다.거의 0이다라고 나온다.
```

```python
# 2차원 -> 3차원으로 변환
def random_projection(X, dimension=3, rseed=42):
    assert dimension >= X.shape[1] # 행,열(2차원) => 차원확대만 가능
    rng=np.random.RandomState(rseed)
    C=rng.randn(dimension, dimension)
    print(C.shape) # 3X3
    print(np.dot(C,C.T), '\n') # 정방, 대칭
    
    e,V = np.linalg.eigh(np.dot(C,C.T)) # 고유치, 고유벡터
    print('V는 : ',V.shape) # 3X3
    print('차원은 : ',V[:X.shape[1]],'\n') # 3차원
    return np.dot(X,V[:X.shape[1]]) # 2X3

print(X.shape)
print(X.shape[1])
print(X.shape)
X3=random_projection(X,3)
print(X3.shape)
```

```python
def make_hello_s_curve(X):
    t=(X[:,0]-2)* 0.75*np.pi
    x=np.sin(t)
    y=X[:,1]
    z=np.sign(t) *(np.cos(t)-1)
    return np.vstack((x,y,z)).T

XS=make_hello_s_curve(X)

ax=plt.axes(projection='3d')
ax.scatter3D(XS[:,0],XS[:,1],XS[:,2], **colorize)
# MDS는 비선형에 취약, 선형에는 강하지만
```

```python
# 4차원 변수가 4개인 iris를 차원축소시키는것
import os # 디렉토리 관리
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
from matplotlib import pylab
from sklearn.manifold import MDS # 다형체 


CHART_DIR='./' # 현제 디렉토리
colors = ['r','g','b'] 
markers=['o',6,'*']

def plot_iris_mds():
    iris = load_iris()
    X = iris.data
    y = iris.target
    fig = pylab.figure(figsize=(10, 4))
    ax = fig.add_subplot(121, projection='3d') # 1,2,1
    ax.set_facecolor('white')
    mds = MDS(n_components=3) # 변수 4 -> 3
    Xtrans = mds.fit_transform(X) # 150 X 3
    
    
    for cl, color, marker in zip(np.unique(y), colors, markers):
        ax.scatter(
            Xtrans[y == cl][:, 0], Xtrans[y == cl][:, 1], Xtrans[y == cl][:, 2], c=color, marker=marker, edgecolor='black')
    pylab.title("3차원에서 iris MDS")
    ax.view_init(10, -15) # 카메라 각도

    # 2차원 축소 3 -> 2
    mds = MDS(n_components=2)
    Xtrans = mds.fit_transform(X)

    ax = fig.add_subplot(122) # 1,2,2
    for cl, color, marker in zip(np.unique(y), colors, markers):
        ax.scatter(
            Xtrans[y == cl][:, 0], Xtrans[y == cl][:, 1], c=color, marker=marker, edgecolor='black')
    pylab.title("2차원에서 iris MDS")

    filename = "mds_demo_iris.png"
    pylab.savefig(os.path.join(CHART_DIR, filename), bbox_inches="tight") 
```

```python
from sklearn.svm import SVC # Classifiction 분류
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler # kmean, PCA는 전처리에 민감하다.

cancer = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0)

scaler=MinMaxScaler().fit(X_train) # 이렇게 연결하는 것을  chaining
X_train_scaled=scaler.transform(X_train)

svm=SVC()
svm.fit(X_train_scaled, y_train)

X_test_scaled=scaler.transform(X_test) # 테스트데이터도 해준다.

print('테스트 점수 : {:.2f}'.format(svm.score(X_test_scaled, y_test)))
```

```python
# 격자 : 그물망
from sklearn.model_selection import GridSearchCV

param_grid={'C':[0.001,0.01,0.1,1,10,100], 
           'gamma':[0.001,0.01,0.1,1,10,100]}

grid=GridSearchCV(SVC(), param_grid=param_grid, cv=5)
grid.fit(X_train_scaled, y_train)
print('최상의 교차 검증 정확도 : {:.2f}'.format(grid.best_score_))
print('테스트 점수 : {:.2f}'.format(grid.score(X_test_scaled,y_test)))
print('최적의 매개변수 : ',grid.best_params_)
```

```python
from sklearn.pipeline import Pipeline

# 참조 : 파라미터 전달 이름을 뭊여준다.
pipe = Pipeline([('scaler', MinMaxScaler()), ('svm',SVC())])
pipe.fit(X_train, y_train)
print('테스트 점수 : {:.2f}'.format(pipe.score(X_test, y_test)))
```

```python
# 격자 : 그물망
# pipeline + GridSearchCV은 다양한 테스트 문제를 해결
from sklearn.model_selection import GridSearchCV

# ragularization 규제 : 과적합 방지
param_grid={'svm__C':[0.001,0.01,0.1,1,10,100], # 처음엔 대충 결정되면 그거에서 좀더 작게
           'svm__gamma':[0.001,0.01,0.1,1,10,100]}

grid=GridSearchCV(pipe, param_grid=param_grid, cv=5)
grid.fit(X_train_scaled, y_train)
print('최상의 교차 검증 정확도 : {:.2f}'.format(grid.best_score_))
print('테스트 점수 : {:.2f}'.format(grid.score(X_test_scaled,y_test)))
print('최적의 매개변수 : ',grid.best_params_)
```

```python
from sklearn.datasets import load_sample_image
from sklearn.utils import shuffle
from time import time
from matplotlib import font_manager,rc

font_name = font_manager.FontProperties(fname='C:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family=font_name)
```

```python
n_colors=64

# 1바이트 = 8비트 = 표현종류 => 256
china=load_sample_image('flower.jpg') # RGB(256컬러 : 1바이트)
# 컬러값 정규화
china=np.array(china, dtype=np.float64)/255
```

```python
w,h,d=original_shape=tuple(china.shape) # 이미지 행x열, 3
print(w,h,d)
assert d==3 # GRB만 들어와라
image_array=np.reshape(china, (w*h,d)) # 2차원으로 변환
print(image_array)
```

```python
# 1000개의 행
image_array_sample=shuffle(image_array, random_state=0)[:1000]
# 64컬러로
kmeans=KMeans(n_clusters=n_colors, random_state=0).fit(image_array_sample) # 중심값결정 64

labels=kmeans.predict(image_array) # 예측 라젤 427 X 640
labels
```

```python
# 64개로 군비화한 array에 색을 입히고 64개의 컬러를 만들어내는 과정
def recreate_image(codebook, labels, w,h): # 64컬러값, 이미지 픽셀수
    d=codebook.shape[1] # 64개의 중심값  64 x 3
    image=np.zeros((w,h,d)) # 원래이미지 사이즈
    label_idx=0
    for i in range(w):
        for j in range(h):
            image[i][j] = codebook[labels[label_idx]]
            label_idx +=1
    return image
```

```python
plt.figure(1)
plt.clf()
ax=plt.axes([0,0,1,1])
plt.axis('off')
plt.title('Original 이미지 (95,615 colors)')
plt.imshow(china)
plt.figure(2)
plt.clf()
ax=plt.axes([0,0,1,1])
plt.axis('off')
# 64컬러
plt.imshow(recreate_image(kmeans.cluster_centers_, labels,w,h))
```

```python
from sklearn.datasets import make_blobs
# 데이터 100개 생성, 군집 하나
X, label = make_blobs(100, centers=1)
kmeans = KMeans(n_clusters=1)
kmeans.fit(X)
distances=kmeans.transform(X) # 각데이터의 중심으로 부터값으로 변환

# rvael() 1차원
# 인덱스를 sort하라 값 : argsort
# :: 2개이면 배수로
# 내림차순으로 변경
# 값을 정렬해서 인덱스로 쓰는 이유는 인덱스로 다른 작업을 하기 위해서
sorted_idx=np.argsort(distances.ravel())[::-1][:5]
# print(sorted_idx)
# print(distances[sorted_idx], sorted_idx)
```

```python
sorted_idx
```

```python
f, ax = plt.subplots(figsize=(7,5))
ax.set_title('Single Cluster')
ax.scatter(X[:,0],X[:,1], label='Points')
ax.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],label='Centroid',color='r')

ax.scatter(X[sorted_idx][:,0], X[sorted_idx][:,1],label='이상치',edgecolors='g', facecolors='none',s=100)

ax.legend(loc='best')
plt.show()
```

```python
# PCA : Principle Component Analysis 주성분 분석
# 모델 입력 전단에서 특징 추출(noise 제거)
# PCA의 결과를 모뎅릐 변수로 추가하면 정확도 상승
import numpy as np
from sklearn.decomposition import PCA
X = np.array([[-1,-1],[-2,-1],[-3,-2],[1,1],[2,1],[3,2]])

# 주성분을 2개로 하라
pca = PCA(n_components=2)
pca.fit(X)
print(pca.explained_variance_ratio_) # 설명력
# 차원축소후에 분석을 하면 좋은점
# noise 제거
# 속도개선
# 차원의 저주(너무 차원이 많으면해결이 안됀다) => 복잡한 문제를 해결
```

```python
print(pca.explained_variance_) # 분산이 큰것이 주성분
print(pca.noise_variance_)
```

```python
# svd : 희소행렬 특징추출 , 비정방행렬인 경우 사용
# randomized : 500X500이상이면
# arpack : 0이 많이 나오니까 0을 줄여서  출력하라
pca =PCA(n_components=2, svd_solver='full') # 'randomized', arpack
# singular value decomposition
pca.fit(X)
print(pca.explained_variance_ratio_)
```

```python
%matplotlib inline
import numpy as np
from sklearn.datasets import fetch_lfw_people
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

people = fetch_lfw_people(min_faces_per_person=20, resize=0.7)
image_shape = people.images[0].shape
print(image_shape) # 87(세로) * 65(가로) , 이미지를 읽을때는 가로x세로, 행x열

fig, axes = plt.subplots(2,5,figsize=(15,8), subplot_kw = {'xticks' : (), 'yticks':()})

for target, image, ax in zip(people.target, people.images, axes.ravel()):
    ax.imshow(image)
    ax.set_title(people.target_names[target])
    
```

```python
print('이미지 사이즈 : {}'.format(people.images.shape))
print('클래스 개수 : {}'.format(len(people.target_names)))
```

```python
mask=np.zeros(people.target.shape, dtype=np.bool)

for  target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] =1

X_people = people.data[mask]
y_people = people.target[mask]
X_people = X_people/255.
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_people, y_people, stratify = y_people, random_state=0)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=100, whiten=True, random_state=0).fit(X_train)

X_train_pca = pca.transform(X_train)
X_test_pca = pca.transform(X_test)
```

```python
fig, axes = plt.subplots(3,5,figsize=(15,12), subplot_kw={'xticks' :(),'yticks':()})

for i , (componenet, ax) in enumerate(zip(pca.components_,axes.ravel())):
    ax.imshow(componenet.reshape(image_shape), cmap='viridis')
    ax.set_title('주성분 {}'.format((i+1)))
```
