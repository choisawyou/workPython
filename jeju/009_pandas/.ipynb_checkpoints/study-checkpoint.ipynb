```python
import re
text = '문의사항이 있으면 032-232-3245으로 연락주시기 바랍니다.'

regex = re.compile(r'\d{3}-\d{3}-\d{4}')

print(regex)

matchobj = regex.search(text) # matching 정보? 를 보여준다.
print(matchobj)
ph = matchobj.group() # match된 str을 보여준다.?
print(ph)
```

```python
import pandas as pd
import numpy as np
import re

# data 타입은 딕셔너리
data={'Dave':'iadslba@naver.com', 'Steve':'steve@gmail.com', 'Rob':'ROB', 'Wes':np.nan}
# data 타입을 시리즈로 바꾼다.
data = pd.Series(data)
print(data)
```

```python
print(data.isnull()) # data value에 null이 포함되 있는가?
print('네이버 : \n',data.str.contains('naver')) # data value문자에 naver과 같은게 있는가?
```

```python
pattern=r'[a-z0-9._%+-].*'#'@[a-z0-9.-].*[a-z]'
matches = data.str.match(pattern, flags=re.IGNORECASE) # 대소문자 불구분
print(matches,'\n')
matches2=data.str.findall(pattern) # 대소문자 구분 정규 표현식에 대문자가 없으면 구분이 안됌
print(matches2)
```

```python
data = {'name':['하늘이','찬호박','우리야','함께가','하성공'],
       'age': [40,50,30,20,70],
       'preScore' : [14,28,39,25,32],
       'postScore':[20,90,55,65,79]}

df= pd.DataFrame(data, columns=['name', 'age','preScore','postScore'])
df
```

```python
print(df.preScore.sum())
print(df.preScore.cumsum()) # 누적 합을 보여준다. 과정까지
print(df.iloc[:,1:].apply(np.mean))
# print(df.iloc[:,1:].map(np.mean)) # dataframe에서는 map을 쓸 수 없다.
# apply()를 쓰자

df.age.mean()
```

```python
# 데이터를 받아오면 데이터를 알아야 한다.
print(df.dtypes,'\n')
print(df.describe())
```

```python
# 정수형으로 인식
df['Customer Number']= df['Customer Number'].astype('int')
df
```

```python
# $ 제거
# 숫자에 문자 제거
df['2016'] = df['2016'].str.replace('$','')
df['2016'] = df['2016'].str.replace(',','')
df
```

```python
# $ 제거
# 숫자에 문자 제거
df['2017'] = df['2017'].str.replace('$','')
df['2017'] = df['2017'].str.replace(',','')
df
```

```python
df.loc[:,['2016','2017'] ] = df.loc[:,['2016','2017'] ].astype(float)
df.describe()
df.dtypes
```

```python
# % 제거
# 숫자에 문자 제거
# df['Percent Growth'] = df['Percent Growth'].str.replace('%','')
# df['Percent Growth'] = df['Percent Growth'].astype('float')
# df['Percent Growth'] = df['Percent Growth'] / 100
# df

def convert_percent(val):
    new_val = val.replace('%','')
    return float(new_val) /100

df['Percent Growth'] = df['Percent Growth'].apply(convert_percent)
# df['Percent Growth'] = df['Percent Growth'].map(convert_percent)
df
```

```python
# Y : 1, N : 0
df.Active = df.Active == 'Y' # 맞다면 True, 아니면 False로 나옴
df
```

```python
# 숫자에 문자 제거
# ignore : 바뀌지않고 타입도 그대로
# raise : If ‘raise’, then invalid parsing will raise an exception
# 아마도 예외 발생??
# coerce : 문자열이 nan로 바뀌고 전체 타입이 float로 바뀜

df['Jan Units'] = pd.to_numeric(df['Jan Units'], errors = 'coerce')
df
```
